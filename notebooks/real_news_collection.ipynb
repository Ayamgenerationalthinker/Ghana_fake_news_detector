```python
fake_df = pd.DataFrame(fake_data)
fake_df.to_csv('../data/fake_news.csv', index=False)
fake_df.head()
```

```python
combined_df = pd.concat([df, fake_df], ignore_index=True)
combined_df.to_csv("../data/full_dataset.csv", index=False)
combined_df.head()
```

```python
import re

# Function to clean raw news text
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"http\S+", "", text)  # Remove URLs
    text = re.sub(r"[^a-zA-Z\s]", "", text)  # Remove punctuation and numbers
    text = re.sub(r"\s+", " ", text).strip()  # Remove extra whitespace
    return text

# Apply cleaning to all articles
combined_df['clean_text'] = combined_df['text'].apply(clean_text)

# View sample cleaned text
combined_df[['title', 'clean_text', 'label']].head()
                                                                                                                                                                                                                                                                                                                                                      
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000)
X = vectorizer.fit_transform(combined_df['clean_text'])

y = combined_df['label'].apply(lambda x: 1 if x == "FAKE" else 0)
```

```python
import os, joblib

# Make sure the 'models' folder exists
os.makedirs("models", exist_ok=True)

# Save vectorizer to file
joblib.dump(vectorizer, "models/tfidf_vectorizer.pkl")
```

```python
from sklearn.model_selection import train_test_split

# 80 % of the data for training, 20 % for testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
print(f"Train size: {X_train.shape[0]}  |  Test size: {X_test.shape[0]}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create and train the model
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print()
print("Classification report:\n", classification_report(y_test, y_pred, target_names=["REAL","FAKE"]))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# === 1. Reload everything =================================================
import joblib, re
from pathlib import Path

# Adjust the path if your notebook location differs
model_bundle_path = Path("../models")
vectorizer = joblib.load(model_bundle_path / "tfidf_vectorizer.pkl")
clf        = joblib.load(model_bundle_path / "logreg_fake_news.pkl")

# === 2. Cleaning helper ===================================================
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)            # strip URLs
    text = re.sub(r"[^a-zA-Z\s]", "", text)        # remove punctuation/digits
    text = re.sub(r"\s+", " ", text).strip()
    return text

# === 3. Classifier with UNVERIFIED threshold ==============================
def classify_with_uncertainty(text, low=0.30, high=0.70):
    """
    Returns dict: {'label': REAL/FAKE/UNVERIFIED, 'prob_fake': 0–1 float}
    """
    cleaned = clean_text(text)
    vec     = vectorizer.transform([cleaned])
    prob    = clf.predict_proba(vec)[0][1]  # prob that it's FAKE
    if prob >= high:
        label = "FAKE"
    elif prob <= low:
        label = "REAL"
    else:
        label = "UNVERIFIED"
    return {"label": label, "prob_fake": round(prob, 3)}

# === 4. Three sample articles to test =====================================
samples = [
    # 4‑A: likely REAL (gov press release style)
    """The Ministry of Education announced on Monday that 1.3 million tablets
    will be distributed to senior high school students beginning August 2025,
    under the Ghana Smart Schools Project launched by President Akufo‑Addo.""",

    # 4‑B: likely FAKE (suspicious giveaway)
    """Breaking! The Ghana Health Service is paying every citizen a free
    200 cedis ‘COVID‑25 bonus’. Click the link below to claim now!""",

    # 4‑C: ambiguous / could be UNVERIFIED
    """Several social‑media posts claim the cedi will be replaced by the ECO
    currency next month, but the Bank of Ghana has yet to confirm.""",
]

for i, txt in enumerate(samples, start=1):
    result = classify_with_uncertainty(txt)
    print(f"Sample {i}:  {result['label']}  (prob_fake={result['prob_fake']})\n")
```
